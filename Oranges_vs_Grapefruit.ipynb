```python
import numpy as np
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
dado = np.loadtxt(url,delimiter = ',',usecols = np.arange(1,6,1),skiprows= 1)
```

```python
diametro_laranja = dado[:5000,0]
peso_laranja = dado[:5000,1]
diametro_toranja = dado[5000:,0]
peso_toranja = dado[5000:,1]
```

```python
def coeficiente_angular(diametro,peso):
  x = diametro
  y = peso
  n = np.size(y)
  a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
  return a

def coeficiente_linear(diametro,peso):
  x = diametro
  y = peso
  n = np.size(y)
  b = np.mean(y) - coeficiente_angular(x , y) * np.mean(x)
  return b
```

```python
reg_laranja = coeficiente_angular(diametro_laranja,peso_laranja) * diametro_laranja + coeficiente_linear(diametro_laranja,peso_laranja)
reg_toranja = coeficiente_angular(diametro_toranja,peso_toranja) * diametro_toranja + coeficiente_linear(diametro_toranja,peso_toranja)
```

```python
plt.plot(diametro_laranja,peso_laranja)
plt.plot(diametro_laranja,reg_laranja, '--r')
```

```python
plt.plot(diametro_toranja,peso_toranja)
plt.plot(diametro_toranja,reg_toranja,'--g')
```

```python
plt.plot(diametro_toranja,peso_toranja)
plt.plot(diametro_laranja,peso_laranja)
plt.title('Oranges_vs_Grapefruit')
```
